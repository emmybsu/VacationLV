# Las Vegas Hotel Ratings

```python
#dependencies and setup

import pandas as pd

#files to load
datafiniti = 'Resources/Datafiniti_Hotel_Reviews.csv'
datafiniti_jun = 'Resources/Datafiniti_Hotel_Reviews_Jun19.csv'
lvtrip_advisor = 'Resources/lasvegas_tripadvisor.csv'
reviews = 'Resources/reviews.csv'


#Read files and store into dataframe

datafiniti_df = pd.read_csv(datafiniti)
datafiniti_jun_df = pd.read_csv(datafiniti_jun)
lvtrip_advisor_df = pd.read_csv(lvtrip_advisor)
reviews_df = pd.read_csv(reviews)
reviews_df.head()



```

```python
#delete extraneous columns from review
del_rev_df = reviews_df.drop(columns=['reviews.text','categories','reviews.userCity','reviews.username','reviews.userProvince','reviews.dateAdded','reviews.doRecommend','reviews.id','address','latitude', 'longitude'])
del_rev_df.head()
```

```python
#create dataframe - vegas specific with related columns
vegas_df = del_rev_df[del_rev_df['city']=='Las Vegas']
vegas_df
```

```python
#merge datafiniti - two datasets
frames = [datafiniti_df, datafiniti_jun_df]
datafin = pd.concat(frames)
datafin.head()
```

```python
#delete extraneous columns from datafin
datafin_del_df = datafin.drop(columns=['keys', 'latitude', 'reviews.text', 'reviews.userProvince', 'reviews.username', 'sourceURLs', 'websites', 'reviews.dateAdded'])
datafin_del_df.head()
datafin_del_df2 = datafin_del_df.drop(columns=['longitude', 'address', 'reviews.sourceURLs','reviews.dateSeen', 'primaryCategories', 'categories', 'dateAdded', 'dateUpdated'])
datafin_del_df2.head()
```

```python
#display only las vegas hotels
#create dataframe - vegas specific with related columns
datafin_vegas_df = datafin_del_df2[datafin_del_df2['city']=='Las Vegas']
datafin_vegas_df
```

```python
#identify dups in dataframe on id column
#By default, for each set of duplicated values, 
#the first occurrence is set on False and all others on True.
dups = datafin.duplicated(subset=['id'])
dups
```
